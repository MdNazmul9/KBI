# Analyzing protein sequences with the ProtParam module.
Protein sequences can be analysed by several tools, based on the ProtParam tools on the Expasy Proteomics Server. The module is part of the SeqUtils package.

Protein Sequence Format
The ProteinAnalysis class takes one argument, the protein sequence as a string and builds a sequence object using the Bio.Seq module. This is done just to make sure the sequence is a protein sequence and not anything else.

```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis
my_seq = "MAEGEITTFTALTEKFNLPPGNYKKPKLLYCSNGGHFLRILPDGTV\
DGTRDRSDQHIQLQLSAESVGEVYIKSTETGQYLAMDTSGLLYGSQTPSEECLFLERL\
EENHYNTYTSKKHAEKNWFVGLKKNGSCKRGPRTHYGQKAILFLPLPV"
analysed_seq = ProteinAnalysis(my_seq)
print(analysed_seq.molecular_weight())

weights = analysed_seq.molecular_weight()
```

```python
print(analysed_seq.gravy())
```

```python
print(analysed_seq.count_amino_acids())
```

```python
print(analysed_seq.get_amino_acids_percent())

sc = analysed_seq.get_amino_acids_percent()
```

print(analysed_seq.)

instability_index: Implementation of the method of Guruprasad et al. (1990, Protein Engineering, 4, 155-161). This method tests a protein for stability. Any value above 40 means the protein is unstable (=has a short half life).

```python
print(analysed_seq.instability_index())
```

```python
print(analysed_seq.flexibility())
```

isoelectric_point: This method uses the module IsoelectricPoint to calculate the pI of a protein.

```python
print(analysed_seq.isoelectric_point())
```

```python
print(analysed_seq.secondary_structure_fraction())
```

```python
Scale = analysed_seq.get_amino_acids_percent()
window= 9
edge =1

print(analysed_seq.protein_scale(Scale, window, edge))
```

```python
print(analysed_seq.protein_scale(analysed_seq.get_amino_acids_percent(), window=9, edge=0.4))
```

```python
import matplotlib.pyplot as plt
rs = analysed_seq.protein_scale(analysed_seq.get_amino_acids_percent(), window=5, edge=0.4)
plt.plot(rs)
plt.ylabel('Protein')
plt.show()
%matplotlib inline
```
